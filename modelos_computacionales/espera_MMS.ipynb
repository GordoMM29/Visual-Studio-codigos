La gerente de una tienda de abarrotes, esta interesada en brindar un buen servicio a las personas mayores que compran en su tienda. Actualmente, la tienda tiene una caja registradora reservada para los clientes de la tercera edad. Esas personas llegan a la caja a un ritmo promedio de 30 clientes por hora, de acuerdo con una distribución de Poisson, y son atendidos a una tasa promedio de 35 clientes por hora, con tiempos de servicios exponenciales.

```python
from math import exp
from math import factorial
```

```python
def UtilizacionPromedio(tasallegada, tasaservicio, servidores):
    return float(tasallegada/(tasaservicio*servidores))

def factor0cioso(tasallegada, tasaservicio, servidores):
    return (100 - (UtilizacionPromedio(tasallegada, tasaservicio, servidores)))

# numero de cientes en cola lq 
def lq(tasallegada, tasaservicio, servidores):
    num = ((Pzero(tasallegada, tasaservicio, servidores)) * (tasallegada/tasaservicio) ** servidores) * UtilizacionPromedio(tasallegada, tasaservicio, servidores)
    den = factorial(servidores) * ((1-UtilizacionPromedio(tasallegada, tasaservicio, servidores)) **2)
    return num/dem

def wq(tasallegada, tasaservicio, servidores):
    return lq(tasallegada, tasaservicio, servidores) / tasallegada

def W(tasallegada, tasaservicio, servidores):
    return wq(tasallegada, tasaservicio, servidores) + (1/tasaservicio)

def L(tasallegada, tasaservicio, servidores):
    return lq(tasallegada, tasaservicio, servidores) + (tasallegada/tasaservicio)
```

```python
#p0
def Pzero(tasallegada, tasaservicio, servidores):
    calc_parte1 = 0
    for x in range(servidores):
        calc_parte1 += ((tasallegada/tasaservicio)**x) / factorial(x)
    
    seccion1 = ((tasallegada/tasaservicio)**servidores) / factorial(servidores)
    
    util = UtilizacionPromedio(tasallegada, tasaservicio, servidores)
    if util >= 1:
        return 0  # Sistema inestable, P0 = 0
    seccion2 = 1 / (1 - util)
    result_seccion2 = seccion1 * seccion2
    
    return 1 / (calc_parte1 + result_seccion2)
```
