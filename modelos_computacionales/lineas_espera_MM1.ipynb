PLANTEAMIENTO:

UNA DULCERIA DE CINE TIENE TRES EMPLEADOS QUE ATIENDEN A LOS CLIENTES POR RIGUROSO ORDEN DE LLEGADA. EL TIEMPO DE SERVICIO POR CLIENTE SE DISTRIBUYE EXPONENCIALMENTE , CON UN PROMEDIO DE 2 MINUTOS POR CLIENTE. LOS CLIENTES DE LA DULCERIA ESPERAN EN UN AMPLIO ESPACIO FORMANDO UNA SOLA FILA Y SUS LLEGADAS TIENEN UNA DISTRIBUCION DE POISSON CON UN PROMEDIO DE 81 CLIENTES POR HORA.

SI EL TIEMPO PROMEDIO QUE PASAN EN EL AREA DE DULCERIA REBASA LOS 10 MINUTOS LOS CLIENTES SE SIENTEN INSATISFECHOS.




```python
# b)
def P(tasaLlegada,tasaServicio):
    return (float(tasaLlegada/tasaServicio)*100)

#numero esperado de clientes en cola (LQ)  d)
def Lq(tasaLlegada,tasaServicio):
    return (float((tasaLlegada*tasaLlegada)/(tasaServicio*tasaServicio-tasaLlegada)))

# numero esperado de clientes recibiendo servicio (LS)  c)
def Ls(tasaLlegada, tasaServicio):
    return tasaLlegada/tasaServicio

# numero esperado de clientes en el sistema de colas (w)  d)
def Lw(tasaLlegada, tasaServico):
    return tasaLlegada/(tasaServicio-tasaLlegada)

# a)
def P0(tasaLlegada, tasaServicio):
    return 1-(tasaLlegada/tasaServicio)

# e)
def W(tasaLlegada, tasaServicio):
    return 1 / (tasaServicio - tasaLlegada)

# f)
def Wq(tasaLlegada, tasaServico):
    return Lq(tasaLlegada, tasaServicio)/tasaLlegada
```

```python
tasaLlegada  = int(input("Ingrese la tasa de llegada(lambda): ")) #30
tasaServicio  = int(input("Ingrese la tasa de servicio(mu): ")) #35

print("La probabilidad de que haya cero clientes en el sistema:", P0(tasaLlegada, tasaServicio)*100)
print("Determine la utilizacion promedio del empleado en la caja registradora:", P(tasaLlegada, tasaServicio))
print("Determine el numero promedio de clientes en el sistema:", Ls(tasaLlegada, tasaServicio))
print("Determine numero promedio de clientes formados en fila:", Lq(tasaLlegada, tasaServicio)*100)
print("Estime el tiempo promedio que los clientes pasan en el sistema:", W(tasaLlegada, tasaServicio), W(tasaLlegada, tasaServicio)*60)
print("Tiempo promedio de espera en fila:", Wq(tasaLlegada, tasaServicio)*100)
```
