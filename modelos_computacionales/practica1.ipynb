```python
# importacion para el uso de librerias
import pandas as pd 
import matplotlib.pyplot as pit
from matplotlib import pyplot as plt
```

```python
# declaracion de datos a utilizar en vectores y matrices
nombres = ['Ana', 'Luis', 'Carlos', 'Marta', 'Sofía']
calificaciones = [[5,5.5,10], [9,8,7], [6,9,8], [7,6,5], [8,7,9]]
```

```python
#crear data frame
df = pd.DataFrame({'nombres': nombres, 'calificaciones': calificaciones})
df
```

```python
#calculos
medias = []
for x in df['calificaciones']:
    medias.append(sum(x) / len(x))
    
df['media'] = medias
cnt = 0
```

```python
df
```

```python
#graficacion
plt.bar(df['nombres'], df['media'], color='red')
plt.xlabel('Estudiantes')
plt.ylabel('promedios')
plt.title('Promedios de calificaciones por estudiante')
for i in df['media']:
    plt.text(cnt, i, str(round(i,2)), ha='center', va='bottom')
    cnt += 1
plt.show()
```
