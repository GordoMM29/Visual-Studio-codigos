```python
# imports 
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargar el csv
df = pd.read_csv('tienda_ropa.csv')
# mostrar todos los datos
print(df)
```

```python
# declarar las varibales para el analisis
x = df['Precio']
y = df['Ventas_Mensuales']
```

```python
# realizar el analisis de regresion lineal 
m, b = np.polyfit(x, y, 1)
# mostrar la ecuacion de la recta
print(f'La ecuacion de la recta es: y = {m}x + {b}')
```

```python
# realizar el coeficiente de correlacion
correlation_matrix = np.corrcoef(x, y)
correlation_xy = correlation_matrix[0,1]
correlation_xx = correlation_matrix[0,0]
correlation_yy = correlation_matrix[1,1]
correlation_coefficient = correlation_xy / np.sqrt(correlation_xx * correlation_yy)
# mostrar el coeficiente de correlacion
print(f'El coeficiente de correlacion es: {correlation_coefficient}')
```

```python
# realizar el coeficiente determinacion 
r_squared = correlation_coefficient ** 2
# mostrar el coeficiente de determinacion
print(f'El coeficiente de determinacion es: {r_squared}')
```

```python
# realizar la ecuacion de la recta 
y_pred = m * x + b
# graficar los datos y la recta de regresion
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, y_pred, color='red', label='Recta de Regresion')
plt.xlabel('Precio')
plt.ylabel('Ventas Mensuales')
plt.title('Regresion Lineal entre Precio y Ventas Mensuales')
plt.legend()
plt.show()
```

```python
# generar una prediccion para un nuevo valor de precio
nuevo_precio = 50
prediccion_ventas = m * nuevo_precio + b
print(f'La prediccion de ventas para un precio de {nuevo_precio} es: {prediccion_ventas}')
```

Analisis de datos del punto 1.

Se realizo un analisis de las ventas en una tienda de ropa declarando asi x y, x siendo el precio de las prendas y siendo las ventas mensuales. Se tiene registro sobre la tienda ya con esto podremos calcular los siguientes datos como seria al analisis de regresion lineal incluyendo una ecuacion de la recta, el coeficiene de correlacion nos dira si el precio de alguna prenda esta relacionado con las ventas mensuales que se tienen en registro, el coeficiente de determinacion nos da una forma de estimar las ventas segun los precios que se tiene en registro y la prediccion sirve para darse una idea que cuando se podria vender o ganar con este tipo de datos para sabe si beneficia o no a la venta de esta tienda.

"Bootstrapping para intervalos de confianza"

```python
# ejemplo bootstrap: media y mediana con datos aleatorios
import numpy as np
import matplotlib.pyplot as plt

# generar datos asimetricos para un ejemplo mas interesante
rng = np.random.default_rng(42)
data = np.exp(rng.normal(loc=3.0, scale=0.7, size=200))

def bootstrap_stat(x, stat=np.mean, n_boot=2000):
    n = len(x)
    boots = np.empty(n_boot)
    for i in range(n_boot):
        sample = rng.choice(x, size=n, replace=True)
        boots[i] = stat(sample)
    return boots

# calcular bootstrap para media y mediana
boots_mean = bootstrap_stat(data, stat=np.mean, n_boot=2000)
boots_median = bootstrap_stat(data, stat=np.median, n_boot=2000)

# estimadores y CIs (percentil 95%)
mean_est = data.mean(); median_est = np.median(data)
mean_ci = np.percentile(boots_mean, [2.5, 97.5])
median_ci = np.percentile(boots_median, [2.5, 97.5])

print(f"Media (orig): {mean_est:.3f}, 95% CI: [{mean_ci[0]:.3f}, {mean_ci[1]:.3f}]")
print(f"Mediana (orig): {median_est:.3f}, 95% CI: [{median_ci[0]:.3f}, {median_ci[1]:.3f}]")

# visualizar distribuciones bootstrap
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(boots_mean, bins=40, color='C0', alpha=0.7)
plt.axvline(mean_est, color='k', label='orig')
plt.title('Bootstrap: media')
plt.legend()
plt.subplot(1,2,2)
plt.hist(boots_median, bins=40, color='C1', alpha=0.7)
plt.axvline(median_est, color='k', label='orig')
plt.title('Bootstrap: mediana')
plt.legend()
plt.tight_layout()
plt.show()
```
