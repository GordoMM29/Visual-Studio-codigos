Los datos se toman cada mes y el analista a construido la siguiente matriz de trancicion de datos historicos.

[G = 0.7 0.2 0.1]
[H = 0.2 0.75 0.05]
[I = 0.1 0.1 0.8]

cuales son los porcentajes de mercado en el estado estable para las dos cervezerias mas grandes?
