```python
#importar librerias
import matplotlib.pyplot as plt
```

```python
# Datos 
ventas = [120, 80, 150, 200, 90, 170, 130]
dias = ["Lun", "Mar", "Mié", "Jue", "Vie", "Sáb", "Dom"]
```

```python
# Calcular promedio 
promedio = sum(ventas) / len(ventas)
```

```python
# Gráfica de barras 
plt.bar(dias, ventas, color='red') 

# Título 
plt.title('Ventas Semanales de la Tienda')

# Línea horizontal con  el promedio 
plt.axhline(y=promedio, color='black', linestyle='--', 
           label=f'Promedio: {promedio:.0f}')

# Etiquetas 
plt.xlabel('Días de la Semana') 
plt.ylabel('Número de Ventas') 
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Mostrar el promedio
print(f"Promedio de ventas calculado: {promedio:.2f}")
```
