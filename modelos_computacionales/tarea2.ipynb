Ejercicio 1

```python
#genrar los numeros randoms entre 0 al 100.
import matplotlib.pyplot as plt
import random 
import numpy as np
import math

x_valores = [random.randint(0, 100) for _ in range(20)]
y_valores = [math.sqrt(x) for x in x_valores]
```

```python
#generar el grafico 
plt.figure(figsize=(10, 6))
plt.scatter(x_valores, y_valores, color='blue', s=100, alpha=0.7, edgecolors='black')

plt.title('Gráfico de Dispersión: Número vs Raíz Cuadrada', fontsize=14)
plt.xlabel('Número (x)', fontsize=12)
plt.ylabel('Raíz Cuadrada (y)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#mostrar los datos
print(f"Números generados (x): {x_valores}")
print(f"Raíces cuadradas (y): {[round(y, 2) for y in y_valores]}")
print("\n")
```

Ejercicio 2

```python
# 2 matrices de 3x3 con valores random
import numpy as np
import random

matriz_a = np.array([[random.randint(-1000, 1000) for _ in range(3)] for _ in range(3)])
matriz_b = np.array([[random.randint(-1000, 1000) for _ in range(3)] for _ in range(3)])
```

```python
producto_escalar = np.dot(matriz_a, matriz_b)
```

```python
print("Matriz A (3x3):")
print(matriz_a)
print("\nMatriz B (3x3):")
print(matriz_b)
print("\nProducto escalar A·B (3x3):")
print(producto_escalar)
print("\n")
```

```python
# Producto escalar de las matrices 3x2 y 2x2
import numpy as np
import random

matriz_c = np.array([[random.randint(-500, 500) for _ in range(2)] for _ in range(3)])
matriz_d = np.array([[random.randint(-500, 500) for _ in range(2)] for _ in range(2)])
```

```python
# Imprimir todos los datos solicitados
print("Matriz C (3x2):")
print(matriz_c)
print("\nMatriz D (2x2):")
print(matriz_d)
producto_escalar_cd = np.dot(matriz_c, matriz_d)
print("\nProducto escalar C·D (3x2):")
print(producto_escalar_cd)
```
