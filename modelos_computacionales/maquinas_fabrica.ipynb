```python
#importar librerias 
import numpy as np
```

```python
#generar matriz de transicion
m1 = np.matrix([[0.80,0.15,0.05],[0.3,0.6,0.1],[0.4,0.3,0.3]])
m1
```

```python
# cual es la probabilidad de que al dia siguiente este en mantenimiento?
# F,M,A
v3 = np.matrix([[1,0,0]])
v3_resultado = np.dot(v3,m1)
v3_resultado
print("La probabilidad de que al dia siguiente este en mantenimiento es: ",v3_resultado[0,1]*100)
```

```python
#4 probabilidad a dos dias de estar en funcionamiento a mantenimiento 
p2 = np.dot(m1,m1)
p2

v4 = np.matrix([[1,0,0]])
v4_resultado = np.dot(v4,p2)
v4_resultado
print("La probabilidad de pasar al dia sigueinte de funioanando a averaido es:", round(v4_resultado[0,2]*100,1))
```

```python
#detectar matriz estable
for i in range(100):
    if i == 0:
        m_aux = np.dot(m1, m1)
        print(m_aux)
        print("1------------------", i + 1)
    else:
        y = i+1
        m_aux_anterior = m_aux
        m_aux = np.dot(m_aux, m1)
        print(m_aux)
        if (np.array_equal(m_aux, m_aux_anterior)):
            break
        print("------------------", i+1)
        print()
        print("la matriz estable es: \n" ,np.round(m_aux,4))

vector_estable = np.round(m_aux[0],4)
#convertir a porcentaje
porcentajes = vector_estable * 100
print("Distribucion estable estacionaria en %")
print(f"funcionando:{porcentajes[0,0]:.2f}")
print(f"mantenimiento:{porcentajes[0,1]:.2f}")
print(f"averiado:{porcentajes[0,2]:.2f}")
```
