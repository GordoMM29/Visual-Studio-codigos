1.- El departamento de estudios de mercado de una fábrica estima que el 20% de la gente que compra un producto un mes, no lo comprará el mes siguiente.

Además, el 30% de quienes no lo compren un mes lo adquirirá al mes siguiente.

En una población de 1000 individuos, 100 compraron el producto el primer mes. ¿Cuántos lo comprarán al mes próximo? ¿Y dentro de dos meses?

```python
#importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Datos
total = 1000
compradores_mes0 = 100
no_compradores_mes0 = total - compradores_mes0

# Probabilidades de transicion
# De los que compran: 80% siguen comprando, 20% dejan
# De los que no compran: 30% compran, 70% siguen sin comprar

# Calcular para el mes 1
compradores_mes1 = (compradores_mes0 * 0.80) + (no_compradores_mes0 * 0.30)
no_compradores_mes1 = total - compradores_mes1  # O también:
# no_compradores_mes1 = compradores_mes0 * 0.20 + no_compradores_mes0 * 0.70

# Calcular mes 2
compradores_mes2 = (compradores_mes1 * 0.80) + (no_compradores_mes1 * 0.30)
no_compradores_mes2 = total - compradores_mes2

# Resultados de las operaciones
print("Mes 0:")
print(f"  Compradores: {compradores_mes0}")
print(f"  No compradores: {no_compradores_mes0}\n")

print("Mes 1:")
print(f"  Compradores: {compradores_mes1:.0f}")
print(f"  No compradores: {no_compradores_mes1:.0f}\n")

print("Mes 2:")
print(f"  Compradores: {compradores_mes2:.0f}")
print(f"  No compradores: {no_compradores_mes2:.0f}")
```

2.- (Cadena de MARKOV) La cervecería más importante del mundo (Guiness) ha contratado a un analista de investigación de operaciones para analizar su posición en el mercado. Están preocupados en especial por su mayor competidor (Heineken). El analista piensa que el cambio de marca se puede modelar como una cadena de Markov incluyendo tres estados, los estados G y H representan a los clientes que beben cerveza producida por las mencionadas cervecerías y el estado I representa todas las demás marcas. Los datos se toman cada mes y el analista ha construido la siguiente matriz de transición de los datos históricos.

(0.7, 0.2, 0.1)
(0.2, 0.75, 0.05)
(0.1, 0.1, 0.8)

```python
P = np.array([
    [0.70, 0.20, 0.10],
    [0.20, 0.75, 0.05],
    [0.10, 0.10, 0.80]
])

# Metodo 1: Resolver sistema πP = π, π1 = 1
Iden = np.eye(3)
A = (P.T - Iden)  

# Construimos matriz aumentada
M = np.vstack([P.T - Iden, [1, 1, 1]]) 
b = np.array([0, 0, 0, 1])

# Resolver por minimos cuadrados 
pi, _, _, _ = np.linalg.lstsq(M, b, rcond=None)

print("Distribución estacionaria:")
print(f"Guiness (G): {pi[0]:.4f}  → {pi[0]*100:.1f}%")
print(f"Heineken (H): {pi[1]:.4f}  → {pi[1]*100:.1f}%")
print(f"Otras (I): {pi[2]:.4f}  → {pi[2]*100:.1f}%")
print(f"Suma: {sum(pi):.4f}")
```

3.- Problema Planteado - Teoria de colas

Una empresa tiene un centro de servicio donde los clientes llegan en promedio 8 clientes por hora (λ\lambda) siguiendo un proceso de distribución de Poisson. Para atender a los clientes, hay tres servidores que trabajan de manera paralela y con un tiempo de servicio de 11 clientes por hora(μ\muμ.) que sigue una distribución exponencial.

Se desea analizar el desempeño del sistema de espera y responder las siguientes preguntas:

1. Tiempo de espera promedio en la fila antes de ser atendido.

2. Tiempo total promedio en el sistema (incluyendo servicio).

3. Número promedio de clientes en el sistema en cualquier instante.

4. Probabilidad de que un cliente tenga que esperar (es decir, todos los servidores están ocupados).

5. Nivel de utilización del sistema (ratio entre la tasa de llegada y la capacidad total de servicio).

```python
#Datos
lamb = 8
mu = 11
c = 3
# Tasa ofrecida 
a = lamb / mu  
rho = a / c    

# Sumatoria de n=0 a c-1
suma = 0
for n in range(c):
    suma += (a ** n) / math.factorial(n)

# Termino especial
termino = (a ** c) / math.factorial(c) * (1 / (1 - rho))

P0 = 1 / (suma + termino)

print(f"P0 = {P0:.4f}")  # Debe dar 0.4816

# Probabilidad de esperar 
P_wait = (a ** c) / math.factorial(c) * (1 / (1 - rho)) * P0

Lq = P_wait * (rho / (1 - rho))

Wq = Lq / lamb

W = Wq + 1/mu

L = lamb * W

print(f"\nResultados:")
print(f"P0 = {P0:.4f}")
print(f"P_wait = {P_wait:.4f}")
print(f"Lq = {Lq:.4f} clientes")
print(f"Wq = {Wq:.4f} horas = {Wq*60:.2f} minutos")
print(f"W = {W:.4f} horas = {W*60:.2f} minutos")
print(f"L = {L:.4f} clientes")
print(f"Utilización ρ = {rho:.4f} = {rho*100:.2f}%")
```
