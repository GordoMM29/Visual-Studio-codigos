5) Suponga que se va a implantar un nuevo método de producción si mediante una prueba de hipótesis se confirma la conclusión de que el nuevo método de producción reduce el costo promedio de operación por hora. </span>

**a) Dé las hipótesis nula y alternativa adecuadas si el costo promedio de producción actual por hora
es $220.**

**b) En esta situación, ¿cuál es el error tipo I? ¿Qué consecuencia tiene cometer este error?**


### a) Hipótesis
- **H₀ (Hipótesis nula):** μ ≥ 220  
- **H₁ (Hipótesis alternativa):** μ < 220

### b) Error tipo I y su consecuencia
**Error tipo I:**  
Implantar el nuevo método creyendo que reduce los costos, cuando en realidad **no los reduce**.

**Consecuencia:**  
Se adopta un método ineficiente, lo que puede causar un **aumento innecesario en los costos operativos**.


```python
import numpy as np
from scipy import stats

costos_nuevo_metodo = [212, 215, 218, 217, 214, 210, 213, 216, 211, 215]

media_hipotetica = 220

t_stat, p_value = stats.ttest_1samp(costos_nuevo_metodo, media_hipotetica)

print("Media de la muestra:", np.mean(costos_nuevo_metodo))
print("Estadístico t:", t_stat)
print("Valor p (una cola):", p_value / 2)  

alpha = 0.05
if (p_value / 2 < alpha) and (t_stat < 0):
    print("Se rechaza H₀: El nuevo método reduce significativamente los costos.")
else:
    print("No se puede rechazar H₀: No hay evidencia suficiente para afirmar que reduce los costos.")
```

6) Considere la prueba de hipótesis siguiente: </span>

$$
\begin{array}{rclcl} \hline
 H_0&:&\mu_1 \leq 25  &   & \\ 
 H_1&:& \mu_1 > 25&   & \\ \hline
\end{array} 
$$

**En una muestra de 40, la media muestral fue 26.4. La desviación estándar poblacional es 6.**

**a) Calcule el valor del estadístico de prueba.**

**b) ¿Cuál es el valor-p?**

**c) Use $\alpha=0.01$, ¿cuál es su conclusión?**

**d) ¿Cuál es la regla de rechazo?, ¿Cuál es su conclusión?**

```python
import math
from scipy.stats import norm

# Datos del problema
n = 40                  # tamaño de la muestra
x_bar = 26.4            # media muestral
mu_0 = 25               # valor bajo H₀
sigma = 6               # desviación estándar poblacional
alpha = 0.01            # nivel de significancia

# a) Cálculo del estadístico Z
z = (x_bar - mu_0) / (sigma / math.sqrt(n))
print("a) Estadístico de prueba Z:", round(z, 4))

# b) Valor-p (prueba de una cola)
p_value = 1 - norm.cdf(z)
print("b) Valor-p:", round(p_value, 4))

# c) Conclusión usando α = 0.01
if p_value < alpha:
    print("c) Se rechaza H₀: Hay evidencia suficiente para decir que μ > 25.")
else:
    print("c) No se puede rechazar H₀: No hay evidencia suficiente para decir que μ > 25.")

# d) Regla de rechazo
z_critico = norm.ppf(1 - alpha)
print("d) Regla de rechazo: Rechazar H₀ si Z >", round(z_critico, 4))
if z > z_critico:
    print("   Como Z >", round(z_critico, 4), "→ Se rechaza H₀.")
else:
    print("   Como Z <=", round(z_critico, 4), "→ No se rechaza H₀.")
```

7) Considere la prueba de hipótesis siguiente: </span>

$$
\begin{array}{rclcl} \hline
 H_0&:&\mu_1 = 15  &   & \\ 
 H_1&:& \mu_1 \neq 15&   & \\ \hline
\end{array} 
$$

**En una muestra de 50, la media muestral fue 14.15. La desviación estándar poblacional es 3.**

**a) Calcule el valor del estadístico de prueba.**

**b) ¿Cuál es el valor-p?**

**c) Use $\alpha=0.05$, ¿cuál es su conclusión?**

**d) ¿Cuál es la regla de rechazo?, ¿Cuál es su conclusión?**

```python
import math
from scipy.stats import norm

# Datos del problema
n = 50                  # tamaño de muestra
x_bar = 14.15           # media muestral
mu_0 = 15               # media bajo H₀
sigma = 3               # desviación estándar poblacional
alpha = 0.05            # nivel de significancia

# a) Estadístico Z
z = (x_bar - mu_0) / (sigma / math.sqrt(n))
print("a) Estadístico Z:", round(z, 4))

# b) Valor-p para prueba bilateral
p_value = 2 * norm.cdf(z) if z < 0 else 2 * (1 - norm.cdf(z))
print("b) Valor-p:", round(p_value, 4))

# c) Conclusión con α = 0.05
if p_value < alpha:
    print("c) Se rechaza H₀: Existe evidencia significativa de que μ ≠ 15.")
else:
    print("c) No se puede rechazar H₀: No hay evidencia suficiente de que μ ≠ 15.")

# d) Regla de rechazo
z_critico = norm.ppf(1 - alpha / 2)
print("d) Regla de rechazo: Rechazar H₀ si |Z| >", round(z_critico, 4))
if abs(z) > z_critico:
    print(f"   Como |Z| = {abs(z):.4f} > {z_critico:.4f} → Se rechaza H₀.")
else:
    print(f"   Como |Z| = {abs(z):.4f} ≤ {z_critico:.4f} → No se rechaza H₀.")
```

8) Considere la prueba de hipótesis siguiente:  </span>

$$
\begin{array}{rclcl} \hline
 H_0&:&\mu_1 \geq 80  &   & \\ 
 H_1&:& \mu_1 < 80&   & \\ \hline
\end{array} 
$$

**Se usó una muestra de 100, la desviación estándar poblacional es de 12. Calcule el valor-p y dé sus conclusiones para cada uno de los resultados muestrales siguientes, Use $\alpha=0.01$ y de una representación grafica del problema usando una sola grafica para todos los incisos**

**a) $\bar{x}=78.5$**

**b) $\bar{x}=77$**

**c) $\bar{x}=75.5$**

**d) $\bar{x}=78.5$**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros conocidos
mu = 80
sigma = 12
n = 100
alpha = 0.01
z_crit = norm.ppf(alpha)  # Punto crítico para una cola (prueba unilateral)

# Datos de medias muestrales
muestras = [78.5, 77.0, 75.5, 78.5]
colores = ['orange', 'deeppink', 'violet', 'deepskyblue']

# Función para calcular z
def calcular_z(x_bar):
    return (x_bar - mu) / (sigma / np.sqrt(n))

# Valores Z para cada media muestral
z_values = [calcular_z(x) for x in muestras]

# Crear distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Z', color='orange')

# Región de rechazo
plt.axvline(z_crit, color='red', linestyle='--', linewidth=2, label='Región de rechazo (Z < {:.2f})'.format(z_crit))

# Marcar puntos de z para cada media muestral
for i, z in enumerate(z_values):
    plt.scatter(z, norm.pdf(z), color=colores[i], s=60, label=f'x̄ = {muestras[i]} → Z = {z:.2f}')

# Etiquetas y leyenda
plt.title('Representación gráfica de las pruebas de hipótesis (α = 0.01)', fontsize=14)
plt.xlabel('Z')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

9) Las declaraciones de impuestos presentadas antes del 31 de marzo obtienen un reembolso que en promedio es de $1056. Considere la población de los declarantes de “última hora” que presentan su declaración los últimos cinco días del periodo para este trámite (normalmente del 10 al 15 de abril). </span>

**a) Un investigador sugiere que la razón por la que estos declarantes esperan hasta los últimos días se debe a que en promedio obtienen un reembolso menor que los que declaran antes del 31 de marzo. Dé las hipótesis apropiadas de manera que el rechazo de  $𝐻_0$  favorezca la sugerencia de este investigador.**

**b) En una muestra de $400$ personas que presentaron su declaración entre el 10 y el 15 de abril, la media de los reembolsos fue  $\$910$ . Por experiencia se sabe que es posible considerar que la desviación estándar poblacional es  $\sigma= \$1600 $ . ¿Cuál es el valor-p?**

**c) Con  $\alpha=0.05$  , ¿cuál es su conclusión?**

```python
# Importar librerías necesarias
import numpy as np
from scipy.stats import norm

# -------------------------------
# Parte a) Formulación de hipótesis
# -------------------------------
print("Parte a) Hipótesis:")
print("H₀: μ ≥ 1056 (el reembolso promedio es mayor o igual a $1056)")
print("H₁: μ < 1056 (el reembolso promedio es menor a $1056)\n")

# -------------------------------
# Parte b) Cálculo del valor-p
# -------------------------------
# Datos del problema
n = 400               # Tamaño de la muestra
x_bar = 910           # Media muestral
sigma = 1600          # Desviación estándar poblacional
mu_0 = 1056           # Media poblacional bajo H₀

# Calcular el estadístico Z
error_estandar = sigma / np.sqrt(n)
Z = (x_bar - mu_0) / error_estandar

# Calcular el valor-p (prueba de cola izquierda)
valor_p = norm.cdf(Z)  # P(Z < z)

print("Parte b) Resultados:")
print(f"Estadístico Z = {Z:.4f}")
print(f"Valor-p = {valor_p:.4f}\n")

# -------------------------------
# Parte c) Conclusión con α = 0.05
# -------------------------------
alpha = 0.05

print("Parte c) Conclusión:")
if valor_p < alpha:
    print(f"Rechazamos H₀ (p = {valor_p:.4f} < α = {alpha}).")
    print("Hay evidencia suficiente para concluir que el reembolso promedio de los declarantes de última hora es menor que $1056.")
else:
    print(f"No rechazamos H₀ (p = {valor_p:.4f} ≥ α = {alpha}).")
    print("No hay evidencia suficiente para concluir que el reembolso promedio de los declarantes de última hora es menor que $1056.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
mu0 = 1056  # media bajo H0
x_bar = 910
sigma = 1600
n = 400
alpha = 0.05

# Cálculo de z y valor-p
z = (x_bar - mu0) / (sigma / np.sqrt(n))
valor_p = norm.cdf(z)

# Gráfica
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Z', color='orange')
plt.axvline(norm.ppf(alpha), color='red', linestyle='--', label=f'Región de rechazo (Z < {norm.ppf(alpha):.2f})')
plt.scatter(z, norm.pdf(z), color='blue', s=80, label=f'Estadístico Z = {z:.2f}')

plt.title('Prueba de hipótesis para reembolsos tardíos (α = 0.05)', fontsize=14)
plt.xlabel('Z')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

10) Reis, Inc., una empresa en Nueva York de investigación sobre bienes raíces, vigila los montos de las rentas de departamentos en Estados Unidos. A mediados de 2002, la renta promedio de un departamento era $\$895$ por mes. Suponga que, según los estudios trimestrales anteriores, es razonable suponer que la desviación estándar poblacional es $\sigma= \$225$. En un estudio reciente, en una muestra de 180 departamentos en todo el país se obtuvieron las rentas (genere la muestra aleatoria de los 180 departamentos)  </span>
    
  **¿ Estos datos muestrales permiten que Reis concluya que la media de la renta actual de departamentos es superior a la media encontrada en 2002?**
  
   
**a) Dé las hipótesis nula y alternativa.**

**b) ¿Cuál es el valor-p?**

**c) Con $\alpha=0.01$, ¿cuál es su conclusión?**

**d) ¿Qué le recomendaría a Reis hacer ahora?**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos del problema
mu_0 = 895        # Renta promedio histórica
sigma = 225       # Desviación estándar poblacional
n = 180           # Tamaño de la muestra
alpha = 0.01      # Nivel de significancia

# a) Generar una muestra aleatoria de 180 rentas con distribución normal
np.random.seed(0)  
muestra = np.random.normal(loc=905, scale=sigma, size=n)
media_muestral = np.mean(muestra)

# b) Calcular el estadístico de prueba Z
z = (media_muestral - mu_0) / (sigma / np.sqrt(n))

# Calcular el valor-p (prueba unilateral hacia la derecha)
p_value = 1 - stats.norm.cdf(z)

# Imprimir resultados
print("Media muestral:", round(media_muestral, 2))
print("Estadístico Z:", round(z, 3))
print("Valor-p:", round(p_value, 4))

# c) Conclusión
if p_value < alpha:
    print("Se rechaza H₀: Hay evidencia de que la renta actual es mayor a la de 2002.")
else:
    print("No se rechaza H₀: No hay evidencia suficiente de que la renta haya aumentado.")

# d) Recomendación
print("Recomendación: Reis debe seguir recolectando datos o esperar más evidencia para confirmar un aumento en la renta.")


# Curva normal
x = np.linspace(mu_0 - 4*sigma/np.sqrt(n), mu_0 + 4*sigma/np.sqrt(n), 1000)
y = stats.norm.pdf(x, mu_0, sigma/np.sqrt(n))

plt.figure(figsize=(10,6))
plt.plot(x, y, label="Distribución Z", color="orange")
plt.axvline(x=mu_0 + stats.norm.ppf(1 - alpha) * (sigma / np.sqrt(n)),
            color='red', linestyle='--', label="Región de rechazo")
plt.axvline(x=media_muestral, color='blue', linestyle='-', label=f"Media muestral = {round(media_muestral, 2)}")
plt.title("Prueba de hipótesis sobre la renta media de departamentos")
plt.xlabel("Media muestral")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```
